```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())

# todo: find better way to describe strategy.. pct allocated to asset, etc.

#------------------------------
#-- 0. Read in Assets
#------------------------------
df = pd.read_csv('D1_assets.csv')

#------------------------------
#-- 0. Assign Strategies
#------------------------------

# buy 1 at start
PID = 'S01'
START_DATE = '2013-01-01' # to handle diff assets start at diff date
PRICE_COL = '"Adj Close"' # SQL needs double quotes on field name w space

BABA_SD = '2014-10-01'
CRM_SD  = '2014-04-01'
TEAM_SD = '2016-02-01'

asset_qty_chg_cols = """
  (CASE 
     WHEN AID='BABA' AND Date = '{baba_sd}' then 1000/"Adj Close" 
     WHEN AID='CRM' AND Date = '{crm_sd}' then 1000/"Adj Close"
     WHEN AID='TEAM' AND Date = '{team_sd}' then 1000/"Adj Close"
     ELSE 0 
   END) as qty_chg
""".format(baba_sd=BABA_SD, crm_sd=CRM_SD, team_sd=TEAM_SD)

q = """SELECT 
      '{pid}' as PID
     ,df.*
     ,{asset_qty_chg_cols}
    FROM df
    WHERE Date >= '{start_date}'
    ORDER BY Date, PID, AID
    ;""".format(pid=PID, start_date=START_DATE, asset_qty_chg_cols=asset_qty_chg_cols)

df = pysqldf(q)

#------------------------------
#-- 0. Output to csv
#------------------------------
# Key: PID AID Date
df.to_csv('S01_Price_Pct.csv',index=False)
```
