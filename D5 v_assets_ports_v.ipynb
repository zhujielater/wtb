```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())

# todo: 

#------------------------------
#-- 0. Read in D3 D4
#------------------------------
ports_assets_df = pd.read_csv('D3_ports_assets.csv')
ports_df = pd.read_csv('D4_ports.csv')

#--------------------------------------
#-- 1. Union
#--------------------------------------
ports_df['AID'] = ports_df['PID']

q = """SELECT {cols} FROM ports_df
       UNION
       SELECT {cols} FROM ports_assets_df
    ;""".format(cols='PID, AID, Date, mv, cost')
df = pysqldf(q)
# ref code: pd.concat([df1, df2])

#------------------------------
#-- 3. Output to csv
#------------------------------
# Key: PID AID Date
df.to_csv('D5_v_assets_ports_v.csv',index=False)
```
