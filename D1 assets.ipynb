```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())

# function: take one AID, read in csv, append AID, return df

def app_aid(stock):
    df = pd.read_csv(stock + '.csv')
    df['AID'] = stock
    return df

stocks = ['TEAM'
         ,'CRM'
         ,'BABA'
         ,'AMZN'
         ,'AAPL'
         ,'FB'
         ,'GOOGL'
         ,'MSFT'
         ,'AYX'
         ,'BRK-B'
         ,'SPY'
         ]

li = []

for stock in stocks:
    df = app_aid(stock)
    li.append(df)

# union
frame = pd.concat(li, axis=0, ignore_index=True)

# save csv
frame.to_csv('D1_assets.csv',index=False)
```
