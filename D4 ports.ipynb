```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())

# todo: find better way to describe strategy.. pct allocated to asset, etc.

#------------------------------
#-- 0. Read in D3 Ports Assets
#------------------------------
ports_assets_df = pd.read_csv('D3_ports_assets.csv')

#--------------------------------------
#-- 1. Summarize mv cost to ports level
#--------------------------------------
q = """SELECT 
      PID, Date
     ,sum({mv_col}) as {mv_col}
     ,sum({cost_col}) as {cost_col}
    FROM 
      ports_assets_df
    GROUP BY PID, Date
    ORDER BY PID, Date
    ;""".format(mv_col='mv',cost_col='cost')
df = pysqldf(q)

#------------------------------
#-- 3. Output to csv
#------------------------------
# Key: PID AID Date
df.to_csv('D4_ports.csv',index=False)
```
