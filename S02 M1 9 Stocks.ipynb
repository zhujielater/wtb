```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())

# todo: find better way to describe strategy.. pct allocated to asset, etc.

#------------------------------
#-- 0. Read in Assets
#------------------------------
df = pd.read_csv('D1_assets.csv')

#------------------------------
#-- 0. Assign Strategies
#------------------------------

# buy 1000 at day 1 of each month
PID = 'S02'
START_DATE = '2013-01-01' # to handle diff assets start at diff date
PRICE_COL = '"Adj Close"' # SQL needs double quotes on field name w space

# find the first 01 day for each stock
q = """SELECT 
      AID
     ,min(Date) as first_01_day
    FROM df
    WHERE strftime('%d', Date) = '01'
    GROUP BY AID
    ;"""

f01_df = pysqldf(q) # not used, to be used later

asset_qty_chg_cols = """
  (CASE WHEN strftime('%d', Date) = '01' then 1000/"Adj Close" ELSE 0 END) as qty_chg
""".format(start_date=START_DATE)

q = """SELECT 
      '{pid}' as PID
     ,df.*
     ,{asset_qty_chg_cols}
    FROM df
    WHERE Date >= '{start_date}'
    ORDER BY PID, AID, Date
    ;""".format(pid=PID, start_date=START_DATE, asset_qty_chg_cols=asset_qty_chg_cols)

df = pysqldf(q)

#------------------------------
#-- 0. Output to csv
#------------------------------
# Key: PID AID Date
df.to_csv('S02 M1 9 Stocks.csv',index=False)
```
