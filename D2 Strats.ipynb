```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())

# todo: find better way to describe strategy.. pct allocated to asset, etc.

#------------------------------
#-- 0. Read in Assets
#------------------------------
df = pd.read_csv('D1_assets.csv')

#------------------------------
#-- 0. Assign Strategies
#------------------------------

# day 1 buy
START_DATE = '2013-05-01'
PRICE_COL = '"Adj Close"' # SQL needs double quotes on field name w space

asset_qty_chg_cols = """
  (CASE WHEN strftime('%d', Date) = '01' then 1000/"Adj Close" ELSE 0 END) as qty_chg
""".format(start_date=START_DATE)

q = """SELECT 
      'Port1' as PID
     ,df.*
     ,{asset_qty_chg_cols}
    FROM df
    WHERE Date >= '{start_date}'
    ORDER BY Date, PID, AID
    ;""".format(start_date=START_DATE, asset_qty_chg_cols=asset_qty_chg_cols)

df = pysqldf(q)

#------------------------------
#-- 0. Output to csv
#------------------------------
# Key: PID AID Date
df.to_csv('D2_strats.csv',index=False)
```
