```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())

# todo: use function to handle a list of assets

#------------------------------
#-- 0. Read in Asset_Daily
#------------------------------
team_df = pd.read_csv('TEAM.csv')
crm_df  = pd.read_csv('CRM.csv')
baba_df = pd.read_csv('BABA.csv')

#------------------------------
#-- 1. Append AID
#------------------------------
def append_pid_aid(AID, asset_df):
    q = """SELECT
          '{AID}' as AID
         ,{asset_df}.*
        FROM
          {asset_df}
        ;""".format(AID=AID, asset_df=asset_df)
    asset_df = pysqldf(q)
    return asset_df
    
team_df = append_pid_aid(asset_df='team_df', AID='TEAM')
crm_df = append_pid_aid(asset_df='crm_df', AID='CRM')
baba_df = append_pid_aid(asset_df='baba_df', AID='BABA')

#------------------------------
#-- 4. Stack Assets
#------------------------------
q = """SELECT * FROM team_df
       UNION
       SELECT * FROM crm_df
       UNION
       SELECT * FROM baba_df
    ;"""
df = pysqldf(q)
df.to_csv('D1_assets.csv',index=False) # Save assets_df
# ref code: pd.concat([df1, df2])
```
